```python
import pandas as pd

dados = pd.read_csv(
    "C:/Users/yasmi/Documents/trainee-repo-template/Sprint 5/Desafio/etapa-1/financeiro_despmensais_2023.csv",
    sep=";",
    encoding="latin1"
)

# validando os CNPJs (removendo aqueles que não possuiam 14 digitos,ex:"153079" ou que havia algum caractere especial, ex:"xxxxxxxxx53") e transformando os inválidos em "None"
def validar_cnpj(cnpj):
    cnpj = str(cnpj)  
    return cnpj.isdigit() and len(cnpj) == 14  

dados["Favorecido_CNPJ"] = dados["Favorecido_CNPJ"].apply(
    lambda x: x if validar_cnpj(x) else None
)

# valores: garantindo que a coluna Data seja do tipo datetime e a coluna Valor como float
dados['Data'] = pd.to_datetime(dados['Data'], format='%Y-%m-%d')
dados['Valor'] = pd.to_numeric(dados['Valor'], errors='coerce')

# texto: removendo espaços extras e padronizamdo para maiusculas
dados['Grupo_Despesa'] = dados['Grupo_Despesa'].str.strip().str.upper()
dados['Elemento_Despesa'] = dados['Elemento_Despesa'].str.strip().str.upper()
dados['Favorecido'] = dados['Favorecido'].str.strip().str.upper()


dados
```

```python
dados.to_csv('dados-tratados.csv', index=False)
```
